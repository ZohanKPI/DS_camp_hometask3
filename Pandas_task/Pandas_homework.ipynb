```python
file_name2 = 'world_bank.csv' 
GDP = pd.read_csv(file_name2, skiprows=4)
GDP
```

```python
replace_dict2 = {
    'Korea, Rep.': 'South Korea',
    'Iran, Islamic Rep.': 'Iran',
    'Hong Kong SAR, China' : 'Hong Kong'
}
GDP.replace({'Country Name': replace_dict2}, inplace=True)

after_replacement2 = GDP[GDP['Country Name'].isin(replace_dict2.values())]

after_replacement2
```

```python
file_name3 = 'scimago_journal.xlsx'
ScimEn = pd.read_excel(file_name3)
ScimEn.head(10)
```

```python
GDP = GDP.rename(columns={'Country Name': 'Country'})

years_of_interest = [str(year) for year in range(2006, 2016)]
GDP = GDP[['Country'] + years_of_interest]
merged_df = energy.merge(GDP, on='Country').merge(ScimEn, on='Country')
top_15_countries = ScimEn['Country'].head(15).tolist()
merged_df = merged_df[merged_df['Country'].isin(top_15_countries)]
pd.options.display.float_format = '{:,.2f}'.format
merged_df
```

```python
merged_df = merged_df.set_index('Country')
```

```python
merged_df
```

```python
desired_columns = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
merged_df = merged_df[desired_columns]
merged_df
```

```python
def answer_one():
    return merged_df
answer_one()
```

```python
def answer_two():
    Top15 = answer_one()
    avgGDP = Top15[years_of_interest].mean(axis=1).sort_values(ascending=False).head(15)
    return avgGDP
answer_two()
```

```python
def answer_three():
    Top15 = answer_one()
    avgGDP = answer_two()
    country_sixth = avgGDP.index[5]
    gdp_change = Top15.loc[country_sixth, '2015'] - Top15.loc[country_sixth, '2006']
    return gdp_change
answer_three()
```

```python
def answer_four():
    Top15 = answer_one()
    Top15['Citation Ratio'] = Top15['Self-citations'] / Top15['Citations']
    country_max_ratio = Top15['Citation Ratio'].idxmax()
    max_ratio = Top15['Citation Ratio'].max()
    result = (country_max_ratio, max_ratio)
    return result
answer_four()
```

```python
def answer_five():
    Top15 = answer_one()
    Top15['Population estimate (in millions)'] = Top15['Energy Supply'] / Top15 ['Energy Supply per Capita']
    third_populous = Top15.sort_values(by='Population estimate (in millions)', ascending=False).iloc[2].name
    return third_populous
answer_five()
```

```python
def answer_six():
    Top15 = answer_one()
    Top15['Citable documents per Capita'] = Top15['Citable documents'] / Top15['Population estimate (in millions)']
    correlation = Top15['Citable documents per Capita'].corr(Top15['Energy Supply per Capita'], method='pearson')
    return correlation
answer_six()
```

```python
def answer_seven():
    Top15 = answer_one()
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    Top15['Continent'] = Top15.index.to_series().map(ContinentDict)
    grouped = Top15.groupby('Continent')
    result_df = grouped['Population estimate (in millions)'].agg(size='size', sum='sum', mean='mean', std='std')
    return result_df
answer_seven()
```
